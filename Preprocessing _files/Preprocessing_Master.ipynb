```python
## Using 15 makes the total to be 622 and for 20 its 573.

import pandas as pd

df = pd.read_csv("data_final_2.csv")
df.columns = ["project"]
rawlist = list(df.project)

items = len(rawlist)

rawlist_shrinked = []

for item in range(items):
    if(len((rawlist[item].split()))<15):
        print(rawlist[item])
    else:
        rawlist_shrinked.append(rawlist[item])
```

```python
# Removing Punctutation and Numbers:

# For punctuation
import string

for item in range(len(rawlist_shrinked)):
    r = str(rawlist_shrinked[item])
    for i in r:
        if i in string.punctuation:
            r = r.replace(i,"")
    rawlist_shrinked[item] = r

# For Removing numbers;

import re
from string import digits

for item in range(len(rawlist_shrinked)):
    s = rawlist_shrinked[item]
    remove_digits = str.maketrans('', '', string.digits)
    s = s.translate(remove_digits)
    rawlist_shrinked[item] = s
```

```python
# Making all text lowercase

for item in range(len(rawlist_shrinked)):
    lower_case = rawlist_shrinked[item].lower()
    rawlist_shrinked[item] = lower_case
```
